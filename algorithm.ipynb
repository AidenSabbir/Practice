```python
#bubble sort
#time O(n^2)
#space O(1)
def bubble(arr):
    sort = False
    while not sort:
            sort = True
            for i in range(len(arr)-1):
                if arr[i] > arr[i+1]:
                    sort = False
                    arr[i],arr[i+1] = arr[i+1],arr[i]
    return(f"Bubble sorted array {arr}")
```

```python
#insertion sort
#time O(n^2)
#space O(1)
def insertion(arr):
    for i in range(1,len(arr)):
        for j in range(i,0,-1):
            if arr[j-1] > arr[j]:
                arr[j-1],arr[j] = arr[j],arr[j-1]
            else:
                break
    return(f"Insertion sorted array {arr}")
```

```python
#selection sort
#time O(n^2)
#space O(1)
def selection(arr):
    for i in range(len(arr)):
        mininum = i
        for j in range(i+1,len(arr)):
            if arr[j] < arr[mininum]:
                mininum = j
        arr[i],arr[mininum] = arr[mininum],arr[i]
    return(f"Selection sorted array {arr}")
```

```python
#Quick sort
#Time O(n log n)
#Space O(n)
def quick_sort(arr):
    n = len(arr)
    if n <= 1:
        return arr 
    pivit = arr[-1]
    L = [i for i in arr[:-1] if i <= pivit]
    R = [i for i in arr[:-1] if i > pivit]
    L = quick_sort(L)
    R = quick_sort(R)

    return L + [pivit] + R

```

```python
# Counting sort
# time O(n + k) where k in the range of data
# space O(k)
from collections import Counter
arr = [2,4,1,5,7,4,0,1,3]
def counting_sort(arr):    
    count = [0] * (max(arr)+1)
    for x in arr:
        count[x] += 1
    i = 0
    for c in range(max(arr)+1):
        while count[c] > 0:
            arr[i] = c
            i += 1
            count[c] -= 1
    return arr
print(counting_sort(arr))
```
