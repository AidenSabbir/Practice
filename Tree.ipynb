```python
class Tree:
    def __init__(self,value,data=None) -> None:
        self.left = None
        self.right = None
        self.value = value
        self.data = data
    def insert(self,value,data=None):
        if value < self.value:
            if self.left is None:
                self.left = Tree(value,data)
            else:
                self.left.insert(value)
        else:
            if self.right is None:
                self.right = Tree(value,data)
            else:
                self.right.insert(value)
    def s_l_h(self):
        if self.left:
            self.left.s_l_h()
        print(self.value)
        if self.right:
            self.right.s_l_h()
    def find(self,target,path=None):
        if not path:
            path = ['head']
        if target < self.value:
            if self.left:
                path.append('left')
                return self.left.find(target,path)
            else:
                return f'{target} is not in Tree Tried path -> {path}'
        elif target > self.value:
            if self.right:
                path.append('right')
                return self.right.find(target,path)
            else:
                return f'{target} is not in Tree Tried path -> {path}'
                
        else:
            return {'data':self,'path':path}

        
tree = Tree(5)
tree.insert(10,{"name":"Aiden",'Age':'21'})
tree.insert(1)
tree.insert(2)
tree.insert(3)
tree.insert(1)
tree.insert(3)
tree.insert(1)
tree.insert(50)
tree.insert(1)
tree.s_l_h()
found = tree.find(10)
if isinstance(found, dict):
    print(found['path'])
else:
    print(found)
```
