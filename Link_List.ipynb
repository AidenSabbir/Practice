```python
# single link list
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
class Link_list:
    def __init__(self):
        self.head = Node()
    def add(self,data):
        new_node = Node(data)
        cur = self.head
        while cur.next != None:
            cur = cur.next
        cur.next = new_node
    def add_after(self,target,data):
        cur = self.head.next
        while cur:
            if cur.data == target:
                new_node = Node(data)
                new_node.next = cur.next
                cur.next = new_node
                return
            cur = cur.next
        print('target not found')

    def remove(self,value):
        prev = self.head
        cur = self.head.next
        while cur:
            if cur.data == value:
                prev.next = cur.next
                return
            else:
                prev = cur
                cur = cur.next
    def length(self):
        l = 0
        cur = self.head.next
        while cur:
            l += 1
            cur = cur.next
        print(f"Length of list is {l}")
    def display(self):
        elem = []
        cur = self.head.next
        while cur:
            elem.append(cur.data)
            cur=cur.next
        print(elem)


```

```python
#double link list
class DNode:
    def __init__(self,data=None):
        self.data = data
        self.next = None
        self.prev = None
class DLink_list:
    def __init__(self):
        self.head = DNode()
    def add(self,data):
        new = DNode(data)
        cur = self.head
        while cur.next is not None:
            cur = cur.next
        cur.next = new
        new.prev = cur
    def remove(self,target):
        cur = self.head.next
        while cur:
            if cur.data == target:
                if cur.prev:
                    cur.prev.next = cur.next
                if cur.next:
                    cur.next.prev = cur.prev
                return
            cur = cur.next

    def display(self):
        elem = []
        cur = self.head.next
        while cur:
            elem.append(cur.data)
            cur=cur.next
        print(elem)



  
```

```python
import timeit
link1 = Link_list()
link2 = DLink_list()
array = []
for i in range(1, 10000):
        array.append(i)
        link1.add(i)
        link2.add(i)
def main():
    link1 = Link_list()
    for i in range(1, 10000):
        link1.remove(i)
def main2():
    link2 = DLink_list()
    for i in range(1, 10000):
        link2.remove(i)
def main3():
    for i in range(1, 10000):
        array.remove(i)
execution_time = timeit.timeit(main, number=1)
execution_time2 = timeit.timeit(main2, number=1)
execution_time3 = timeit.timeit(main3, number=1)


print(f'Time for single linked list: {execution_time:.5f} seconds')
print(f'Time for double linked list: {execution_time2:.5f} seconds')
print(f'Time for array list: {execution_time3:.5f} seconds')
```
