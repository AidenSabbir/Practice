```python
import random
import string
from collections import defaultdict
names = ['james','aiden','alif','rufa']
letters = string.ascii_letters + string.ascii_lowercase + string.ascii_uppercase + string.digits + ' '
dic = defaultdict()
for name in names:
    dic[name] = ''.join(random.choice(letters) for _ in range(225))
print(dic)
```
